#  Linear Regression - Step-by-Step

##  1. Introduction


Linear regression is a statistical and machine learning method that builds a model (a formula) to 
        Find a relationship between two variables.
        Predict one value (like weight) from another (like height).
It does this by drawing the best-fitting straight line through the data.

It is just a smart way to draw a straight line that helps us guess one thing from another.
Like:
How tall someone is  maybe guess their weight
How many hours you study  maybe guess your score
How much you water a plant  maybe guess how tall it grows


**Linear regression** is a supervised learning algorithm used for **predicting continuous values**. The value we are trying to predict is called dependent variable and the value we are using to predict is called independent variable. 

The equation which defines the simplest form of the regression equation with one dependent and one independent variable: `y = mX+b`.

Where `Y` is Dependent Variable
`X` is Independent Variable
`b` is intercept(mnemonic : b means where the line begins)
`m` is slope (mnemonic : m means move)





###  Real-world examples:
- Predicting house prices based on size, location, etc.
- Estimating sales based on advertising spend
- Predicting temperature based on historical data

```python
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn import linear_model

data = pd.read_csv("../datasets/weight-height.csv")
data.sample(10)

plt.scatter(data["Height"][:100], data["Weight"][:100], color="red")
plt.title("Height Weight Data")
plt.xlabel("Height")
plt.ylabel("Weight")
```

```python
x = data[["Height"]]
y = data["Weight"]
model = linear_model.LinearRegression()
model.fit(x, y)
```

```python
y_pred = model.predict(x)
plt.scatter(data["Height"], data["Weight"], color="blue", label = "Actual Data")
plt.plot(data["Height"], model.predict(data[["Height"]]), color="red", linewidth=2, label="fit line")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Linear Regression - Height vs Weight")
```

```python
m = model.coef_
b = model.intercept_
X = 80
y = (m*X) + b
y
```
